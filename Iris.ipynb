```python
import pandas as pd
```

```python
#importando o dataset
from sklearn.datasets import load_iris

#retornando os dados
iris = load_iris()
iris
```

```python
#transformando em dataframe
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df
```

```python
#selecionando apenas as colunas de pétalas
iris1 = iris_df.loc[iris_df.target.isin([1,2]),['petal length (cm)', 'petal width (cm)', 'target']]
iris1
```

```python
#separando x e y
x = iris1[['petal length (cm)', 'petal width (cm)']]
y = iris1.target
```

```python
#fazendo o train_test_split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
#visualizando os dados de treino
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

ax.scatter(x = x_train['petal width (cm)']
           , y = x_train['petal length (cm)']
           , c = y_train, cmap='viridis')

ax.set(xlim=(0.9,2.6), xticks=[1,1.5,2,2.5],
       ylim=(3,7), yticks=[3,4,5,6,7])

plt.show()
```

```python
#importando o k-NN
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Criando o classificador
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
#fazendo o fit com os dados de treino
clf = clf.fit(x_train, y_train)
```

```python
#fazendo a previsão para os dados de teste
y_pred = clf.predict(x_test)
```

```python
#verificando a matriz de confusão
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
#visualizar os dados de treino e teste
fig,ax = plt.subplots()

ax.scatter(x = x_train['petal width (cm)']
           , y = x_train['petal length (cm)']
           , c = y_train, alpha =0.7,
           cmap='viridis')

ax.scatter(x = x_test['petal width (cm)']
           , y = x_test['petal length (cm)']
           , c = y_pred, alpha =0.2,
           cmap='RdYlGn')

ax.scatter(x = x_test['petal width (cm)']
           , y = x_test['petal length (cm)']
           , c = y_test, alpha =0.2,
           cmap='RdYlGn')

ax.set(xlim=(0.9,2.6), xticks=[1,1.5,2,2.5],
       ylim=(3,7), yticks=[3,4,5,6,7])

plt.show()
```
